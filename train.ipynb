```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/diabetic.txt')
df.info()
```

```python
df['Dstats'] = np.where(df['Diabetic Status']=='Yes',1,0)
df['Dstats']
```

```python
xdf = df.drop(['Diabetic Status','Dstats'], axis = 1)
xdf.head()
```

```python
X = xdf.to_numpy()
X
```

```python
y = df['Dstats'].to_numpy()
y
```

```python
y.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X,y)
```

```python
from sklearn.metrics import r2_score
ycap = model.predict(X)
ycap
```

```python
np.c_[y,ycap]
```

```python
acc = r2_score(y,ycap)
acc
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR as svr

model2 = svr(kernel = 'linear')
model2.fit(X,y)
ycap2 = model2.predict(X)
ycap2
```

```python
np.c_[y,ycap2]
```

```python
acc = r2_score(y,ycap2)
acc
```

```python
from fastapi import FastAPI
from pydantic import BaseModel
```

```python
app = FastAPI()
model = joblib.load("diabetics_model.pkl")
```

```python
class DiabetiesInput(BaseModel):
    Age	:int
    BMI	: float
    BloodPressure : float
    GlucoseLevel	: float
    InsulinLevel : float

@app.get("/")
def read_root():
    return {"Message": f"DiabeticPredictionAPI is live"}

@app.post('/predict')
def predict_diabetics(data: DiabetiesInput):
    input_data = np.array([data.Age, data.BMI, data.BloodPressure, data.GlucoseLevel, data.InsulinLevel])
    prediction = model.predict(input_data.reshape(1, -1)) # Reshape for single prediction
    return {"Prediction": prediction.tolist()[0]} # Convert numpy array to list for JSON response
```
